```python
#BLOCO DE FUNÇÕES-->
#Nesse primeiro bloco, foram definidas as funções a serem chamadas no programa principal.

#A função opções, Mostra as opções que aparecem ao usuário escolher a opção principal
# Nele é recolhido o valor da opção que o usuário deseja realizar de no programa principal, constendo uma parte gráfica
# e um input para o mesmo inserir o número da opçaõ desejada. Nessa função, a opção de escolha inserida pelo usuário,
#também é validada, caso o usuário inserir um número de opção fora do escopo de opções existentes, retornando uma mensa-
#gem de erro.
def opcoes():
    print('-' * 40)
    print('1--> Cadastrar Peças')
    print('2--> Consultar Peças')
    print('3--> Remover Peças')
    print('4--> Sair')
    print('-' * 40)
    while True:
        valor=input('Digite a opção desejada: ')
        if valor=='1' or valor=='2' or valor=='3' or valor=='4':
            return valor
        else:
            print ('Opção inexistente...Digite novamente.')

#Na função cadastrarPeca, é realizado inicialmente a recolha do tamanho da lista de códigos, para que seja gerado o número
#do código da peça a ser cadastrada, inserindo um número a mais da quantidade existente,e mostrado ao cliente o código.
#gerado. Então é pedido pro usuário digitar o nome da peça, o fabricante e o valor, sendo esses, guardados nas respectivas
#listas do dicionário de peças.
def cadastraPeca():
    codigo=1
    for i in range (len(pecas['código'])):
        codigo=int(pecas['código'][i])+1
    codigo=str(codigo)
    print ('Código da Peça: {}'.format(codigo))
    nome=input('Digite o nome da peça: ')
    fabricante = input('Digite o fabricante da peça: ')
    valor=input('Digite o valor da peça: ')
    pecas['código'].append(codigo)
    pecas['nome'].append(nome)
    pecas['fabricante'].append(fabricante)
    pecas['valor'].append(valor)

#A função consultarPeca, Mostra as opções que aparecem ao usuário escolher conultar peças.
# Nele é recolhido o valor da opção que o usuário deseja realizar ao consultar peças, constendo uma parte gráfica
# e um input para o mesmo inserir o número da opçaõ desejada. Nessa função, a opção de escolha inserida pelo usuário,
#também é validada, caso o usuário inserir um número de opção fora do escopo de opções existentes, retornando uma mensa-
#gem de erro.
#Nessa função, também é averiguada a opção escolhida pelo usuário, através de if, e chamada a função relativa à escohida,
#que é comentada mais a frente.
def consultarPeca():
    while True:
        print('-' * 40)
        print('1--> Consultar Todas as Peças')
        print('2--> Consultar Peças pelo Código')
        print('3--> Consultar Peças pelo Fabricante')
        print('4--> Consultar Peças pelo Nome')
        print('5--> Voltar')
        print('-' * 40)
        opcao=input('Digite a Oção Desejada:')
        if opcao=='1':
            consultarTodas()
        elif opcao=='2':
            consultaCodigo()
        elif opcao == '3':
            consultaFabricante()
        elif opcao == '4':
            consultaNome()
        elif opcao == '5':
            break
        else:
            print ('Opção inexistente...Digite novamente.')

#Na função consultarTodas, é chamada pelo consultarPeca, quando o usuário escolher a opção de consultar todas as peças.
#Nessa função, é recolhido o conteúdo da lista contendo o código do produto, nome do produto, fabricante e preço, através
#de um for, onde é utilizado como range, o tamanho da lista interna do dicionário.
def consultarTodas():
    for i in range (0,(len( pecas['código']))):
        print ('-'*40)
        print ('Código: {}'.format(pecas['código'][i]))
        print ('Produto: {}'.format(pecas['nome'][i]))
        print ('Fabricante: {}'.format(pecas['fabricante'][i]))
        print ('Preço: {}'.format(pecas['valor'][i]))
        print('-' * 40)

#Na função consultarTodas, é chamada pelo consultarPeca, quando o usuário escolher a opção de consultar peças por código.
#Nessa função, é recolhido o conteúdo da lista contendo o código do produto, nome do produto, fabricante e preço, através
#de um for, onde é utilizado como range, o tamanho da lista interna do dicionário, e comparado por um if com o código
#digitado pelo usuário através de um input. Ao identificar o código, quando igual ao presente na lista de códigos do dicio-
#nário, então é impresso ao usuário o código do produto, nome do produto, fabricante e preço.
#Nessa função, a opção de código inserida pelo usuário, também é validada, caso o usuário inserir um código
# não existente, retornando uma mensagem de erro.
def consultaCodigo():
    valid=''
    codigo=input('Código do Produto: ')
    for i in range (0, (len(pecas['código']))):
        if pecas['código'][i]==codigo:
            print ('-'*40)
            print ('Código: {}'.format(pecas['código'][i]))
            print ('Produto: {}'.format(pecas['nome'][i]))
            print ('Fabricante: {}'.format(pecas['fabricante'][i]))
            print ('Preço: {}'.format(pecas['valor'][i]))
            print ('-'*40)
            valid = 'ok'
    if not valid:
        print('Produto inexistente...')

#Na função consultarFabricante, é chamada pelo consultarPeca, quando o usuário escolher a opção de consultar peças por fabricante.
#Nessa função, é recolhido o conteúdo da lista contendo o código do produto, nome do produto, fabricante e preço, através
#de um for, onde é utilizado como range, o tamanho da lista interna do dicionário, e comparado por um if com o fabricante
#digitado pelo usuário através de um input. Ao identificar o fabricante, quando igual ao presente na lista de fabricantes do dicio-
#nário, então é impresso ao usuário o código do produto, nome do produto, fabricante e preço.
#Nessa função, a opção de fabricante inserida pelo usuário, também é validada, caso o usuário inserir um fabricante
# não existente, retornando uma mensagem de erro.
def consultaFabricante():
    fabricante=input('Fabricante: ')
    fabricante=(fabricante.lower())
    valid=''
    for i in range (0, len(pecas['fabricante'])):
            listafab=pecas['fabricante'][i]
            listafab=(listafab.lower())
            if  listafab==fabricante:
                print ('-'*40)
                print ('Código: {}'.format(pecas['código'][i]))
                print ('Produto: {}'.format(pecas['nome'][i]))
                print ('Fabricante: {}'.format(pecas['fabricante'][i]))
                print ('Preço: {}'.format(pecas['valor'][i]))
                print('-' * 40)
                valid='ok'
    if not valid:
         print('Produto inexistente...')

#Na função consultarNome, é chamada pelo consultarPeca, quando o usuário escolher a opção de consultar peças por nome.
#Nessa função, é recolhido o conteúdo da lista contendo o código do produto, nome do produto, fabricante e preço, através
#de um for, onde é utilizado como range, o tamanho da lista interna do dicionário, e comparado por um if com o nome
#digitado pelo usuário através de um input. Ao identificar o nome, quando igual ao presente na lista de nomes do dicio-
#nário, então é impresso ao usuário o código do produto, nome do produto, fabricante e preço.
#Nessa função, a opção de nome de peça inserida pelo usuário, também é validada, caso o usuário inserir um nome de peça
#não existente, retornando uma mensagem de erro.

def consultaNome():
    produto=input('Nome do produto:')
    produto=(produto.lower())
    valid=''
    busca=0
    for i in range (0, len(pecas['nome'])):
            listaprod=pecas['nome'][i]
            listaprod=(listaprod.lower())
            busca=listaprod.count(produto)
            if busca:
                print ('-'*40)
                print ('Código: {}'.format(pecas['código'][i]))
                print ('Produto: {}'.format(pecas['nome'][i]))
                print ('Fabricante: {}'.format(pecas['fabricante'][i]))
                print ('Preço: {}'.format(pecas['valor'][i]))
                print('-' * 40)
                valid='ok'
    if not valid:
         print('Produto inexistente...')


#Na função removerPecas, é recolhido o conteúdo da lista contendo o código do produto, nome do produto, fabricante e preço,
# através de um for, onde é utilizado como range, o tamanho da lista interna do dicionário, e comparado por um if com o
#código digitado pelo usuário através de um input. Ao identificar o código, quando igual ao presente na lista de nomes do dicio-
#nário, então é removido o código do produto, nome do produto, fabricante e preço da lista, e retornado uma mensagem.
#Nessa função, a opção de nome de peça inserida pelo usuário, também é validada, caso o usuário inserir um nome de peça
#não existente, retornando uma mensagem de erro.
def removerPecas ():
    codigo = input('Digite o código da peça a ser removida: ')
    valid=''
    for i in range(0, len(pecas['código'])):
        if pecas['código'][i] == codigo:
            num=i
            valid='ok'

    if valid:
        pecas['código'].pop(num)
        pecas['nome'].pop(num)
        pecas['fabricante'].pop(num)
        pecas['valor'].pop(num)
        print('Produto removido com sucesso!')
    else:
        print('Produto inexistente...')

#BLOCO DE LINGUAGEM PRINCIPAL-->
#Inicialmente é definido o dicionário 'pecas' com suas respectivas listas: uma de código de produtos, uma com o nome de produto,
#uma com o fabricante do produto e outra com o valor.
pecas={'código':[],'nome':[],'fabricante':[],'valor':[]}

#Nesse bloco, foi realizado a parte gráfica de mensgaem de boas vindas através de print.
bemvindo='Bem Vindo(a) ao Controle de Estoque da Bicicletaria da Patrícia Canossa Gagliardi'
tam=(len(bemvindo))
tam=90-tam
tam=int(tam/2)
print ('*','-'*90,'*')
print ('|',' '*(tam-1),'{}'.format(bemvindo),' '*(tam), '|')
print ('*','-'*90,'*')

#Após, é iniciado um while, onde dentro dele é chamado a função de opções, onde irá exibir as opções de escolha do programa,
# e então, através de if, veirificado qual opção escolhida pelo usuário, chamando a função relativa a escolha:
#cadastrarPeca no caso do usuário digitar opção 1, consultarPeca caso do usuário digitar opção 2, removerPeca caso do usuário
# digitar opção 3, e terminar o while com break e mostrar mensagem de fim do programa, caso do usuário digitar opção 4.
while True:
    opcao=opcoes()
    if opcao=='1':
        cadastraPeca()
    elif opcao=='2':
        consultarPeca()
    elif opcao=='3':
        removerPecas()
    elif opcao=='4':
        print ('Finalizando programa...')
        break

```
