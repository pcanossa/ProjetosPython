```python
#Bloco de funções-> Nesse primeiro bloco, são definidas as funções que serão chamadas na programação principal
#Na função dimensoesObjeto, é realizado a gravação dos valores de entrada do comprimento, largura e altura do objeto,
#através do uso da função valid(que será descrita mais a frente), nas variáveis 'comprimento', 'largura', e 'altura',
#e realizado então o cálculo do volume na variável 'vol', através dos valores de alrgura, altura e comprimento obtidos.
#Nesse bloco, também é realizado a recolha do preço do volume do objeto, através da função for, para varrer um
#dicionário com listas com os valores pré definidos do multiplicador em relação aos volumes (que será descrito mais a
# frente), e por fim, retornando um tupla  com os valores de volume e do multiplicadornas variáveis 'vol' e 'multiplica-
# dorVolume'.
def dimensoesObjeto():
    print('Digite o comprimento em cm do objeto: ')
    comprimento=valid()
    print('Digite a largura em cm do objeto: ')
    largura=valid()
    print('Digite a altura em cm do objeto: ')
    altura=valid()
    multiplicadorVolume=10
    vol=comprimento*altura*largura
    for i in range (len(volumes['volume'])):
        if vol>=int(volumes['volume'][i]):
            multiplicadorVolume=int(volumes['multiplicador'][i])
    return vol, multiplicadorVolume

#Na função peso objeto, é realizado a gravação dos valores de entrada do peso do objeto, através do uso da função
# valid(que será descrita mais a frente), nas variável 'peso'.
#Nesse bloco, também é realizado a recolha do multiplicador do peso objeto, através da função for, para varrer um
#dicionário com listas com os valores pré definidos do multiplicador em relação aos pesos (que será descrito mais a
# frente), e retornado um tupla com os valores de peso e multiplicador através das variáveis 'peso' e 'multiplicadorPeso'.
def pesoObjeto():
    print('Digite o peso em kg do objeto: ')
    peso=valid()
    print ('O peso do produto é de {}kg\n'.format(peso))
    multiplicadorPeso=1
    for i in range (len(pesos['peso'])):
        if peso >= float(pesos['peso'][i]):
            multiplicadorPeso=float(pesos['multiplicador'][i])
    return peso, multiplicadorPeso

#Na função peso objeto, é realizado a gravação dos valores de entrada da rota do objeto, através do uso da função
# validObjeto (que será descrita mais a frente), nas variável 'trajeto'.
#Nesse bloco, também é realizado a recolha do multiplicador da rota do objeto, através da função for, para varrer um
#dicionário com listas com os valores pré definidos do multiplicador em relação às rotas (que será descrito mais a
# frente)e retornado um tupla com os valores de trajeto e multiplicador através das variáveis 'trajeto' e 'multiplicadorRotas'.
def rotaObjeto():
    print ('Digite a rota desejada: ')
    multiplicadorRotas=''
    for i in range (len(rotas['trajeto'])):
        print ('{} ----> {}'.format(rotas['opções'][i],rotas['trajeto'][i]))
    trajeto=validObjeto()
    for i in range (len(rotas['trajeto'])):
        if trajeto==rotas['opções'][i]:
            multiplicadorRotas=float(rotas['multiplicador'][i])
    return trajeto, multiplicadorRotas

#Na função validObjeto, é realizado a função input para que o usuário insira o valor do peso do objeto e guarado na
# variável 'valor', e então validado #através da função while e for, utilizando o tamanho da lista de trajetos,
# para varrer os valores e comparar com o valor inserido pelo usuário, retornando um uma mensagem de erro
# caso o usuário inserir valor diferente das rotas existentes de escolha, e repetindo o input.
# Nessa função, também é utilizada a função upper na variável 'valor', para tornar maiúsculo o valor digitado, caso o
#usuário tenha digitado em minúsculo.
def validObjeto():
    while True:
        valor = input('-->')
        valorup=(valor.upper())
        for i in range (len(rotas['trajeto'])):
            if valor==rotas['opções'][i] or valorup==rotas['opções'][i]:
                return valor.upper()
                break

        print('Rota não existente! Digite novamente:')


#Na função validObjeto, é realizado a função input para que o usuário insira o valor do guarado na  variável 'valor',
# e então validado através da função while, que através da função try e if, verifica se o valor digitado é um número,
#e não é 0, retornando uma mensagem de erro em caso dessas duas situações, e reornando o valor da variável 'valor'
def valid():
     while True:
        try:
            valor = float(input('-->'))
            if valor !=0:
                break
            else:
                print('O valor não pode ser igual a 0!\n')
        except:
            print ('Valor incorreto, digite novamente...\n')
     return valor

#Na função calculavolume, é chamado a função de dimensoesObjeto e retornado através de print o volume do objeto ao
#usuário, e retornado a tupla gerada pela função dimensoesObjeto.
def calculavolume():
    global volumev
    volumev=0
    volumev = (dimensoesObjeto())
    print('O volume do objeto é de {:.2f}cm³\n'.format(volumev[0]))
    return volumev


#Bloco de programação principal-->
#Inicialmente é realizada a parte gráfica de mesagem de boas vindas ao usuário através da função print
bemvindo='Bem Vindo a Companhia Logística de Patrícia Canossa Gagliardi'
tam=(len(bemvindo))
tam=80-tam
tam=int(tam/2)
print ('*','-'*80,'*')
print ('|',' '*(tam-1),'{}'.format(bemvindo),' '*(tam), '|')
print ('*','-'*80,'*')

#Nessa parte, foram definidos os dicionários com listas de valores pré definidos
volumes={'volume':['0', '1000', '10000', '30000'],'multiplicador':['10','20','30','50']}
#no dicionário volumes, foi elaborado uma lista com o valor mínimo de volumes para cada categoria de multiplicador, e
#outra com os valores dos multiplicadores.
pesos={'peso':['0','0.1','1','10'],'multiplicador':['1','1.5','2','3']}
#no dicionário pesos, foi elaborado uma lista com o valor mínimo de pesos para cada categoria de multiplicador, e
#outra com os valores dos multiplicadores.
rotas={'opções':['PC','SC','AR','BR','RP'],'trajeto':['De São Paulo até Piracicaba','De São Paulo até São Carlos','De São Paulo até Araraquara','De São Paulo até Bauru','De São Paulo até Ribeirão Preto'],'multiplicador':['1','1','1.2','1.2','1.5']}
#no dicionário rotas, foi elaborado uma lista com as opções a serem escolhidas pelo usuário, uma com a descrição das rotas, e
#outra com os valores dos multiplicadores.

#Então é chamada a função calculavolume para realizar a recolha dos valores para cálculo do volume e seu multiplicador
volume = (calculavolume())
fim=''

#Após, foi realizado a validação do valor máximo através da função while e if, verificando se o valor do volume digitado pelo
#usuário é maior ou igual a 100000, e retornando mensagem de erro caso for, sendo pedido também para o usuário informar
#se deseja encerrar o programa ou deseja inserir outro valor.
while True:
    if volume[0]>=100000:
        print ('Não aceitamos transporte de objetos com volumes acima da 100000cm³')
        continuav = input('Deseja calcular novamente (s/n): ')
        if continuav=='s':
            volume = (calculavolume())
        elif continuav=='n':
            print ('Encerrando o programa...')
            fim='ok'
            break
        else:
            print ('Opção inválida...\n')
    else:
        break

#Então, caso o usuário inserir um valor válido ou optar em inserir outro valor, e esse for menor que o valor máximo (100000),
# é chamada a função pesoObjeto para realizar a recolha dos valor do peso e seu multiplicador.
if not fim:
    peso=(pesoObjeto())

 # Após, foi realizado a validação do valor máximo através da função while e if, verificando se o valor do volume digitado pelo
 # usuário é maior que 30, e retornando mensagem de erro caso for, sendo pedido também para o usuário informar
 # se deseja encerrar o programa ou deseja inserir outro valor.
    while True:
        if peso[0]>30:
            print('Não aceitamos transporte de objetos com peso acima da 30kg')
            continuap=input('Deseja calcular novamente (s/n): ')
            if continuap == 's':
                peso=(pesoObjeto())
            elif continuap=='n':
                print ('Encerrando o programa...')
                fim='ok'
                break
            else:
                print ('Opção inválida...\n')
        else:
            break
# Então, caso o usuário inserir um valor válido ou optar em inserir outro valor, e esse for menor que o valor máximo (30),
#É realizado o cáluculo do valor do frete através da multiplicado do valor multiplicador do volume*multiplicador do peso*
#multiplicador da rota, guardado na variável 'frete', e realizada a parte gráfica de impressão, através de print, do valor
#do frete ao usuário e de como foi composto (valor do volume, peso e rota).
    if not fim:
        rota = (rotaObjeto())
        frete=volume[1]*peso[1]*rota[1]
        print('')
        print ('')
        print ('*','-'*80,'*')
        print ('|','Valor do Volume do Objeto: R${:.2f}'.format(volume[1]),' '*(80-len('Valor do Volume do Objeto: R${:.2f}')), '|')
        print ('|','Valor do Peso do Objeto: R${:.2f}'.format(((peso[1])*(volume[1]))-volume[1]),' '*(80-len('Valor do Peso do Objeto: R${:.2f}')), '|')
        print ('|','Valor da Rota do Objeto: R${:.2f}'.format(frete-((volume[1])+((peso[1])*(volume[1])-volume[1]))),' '*(81-len('Valor da Rota do Objeto: R${:.2f}')), '|')
        print ('*','-'*80,'*')
        print ('|','VALOR TOTAL: R${:.2f}'.format(frete),' '*(80-len('VALOR TOTAL: R${:.2f}')), '|')
        print ('*','-'*80,'*')
```
