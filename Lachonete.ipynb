```python
#Bloco de funções: Inicalmente são definidas as funções que realizarão os procedimentos dentro do programa inicial

#Nessa função, chamada menu, é realizado a parte gráfica do menu
def menu():
    print('***************************************CARDÁPIO********************************************************')
    print('    |    CÓDIGO    |                     DESCRIÇÃO                       |   VALOR   |                 ')
    print('    |      100     |              Cachorro Quente Simples                |  R$ 9,00  |                 ')
    print('    |      110     |              Cachorro Quente Duplo                  |  R$ 11,00 |                 ')
    print('    |      120     |                    X Salada                         |  R$ 13,00 |                 ')
    print('    |      130     |                      X Egg                          |  R$ 12,00 |                 ')
    print('    |      140     |                     X Bacon                         |  R$ 14,00 |                 ')
    print('    |      150     |                     X Tudão                         |  R$ 17,00 |                 ')
    print('------------------------------------------BEBIDAS------------------------------------------------------')
    print('    |      200     |                Refrigerante Lata                    |  R$ 5,00  |                 ')
    print('    |      210     |                    Chá Gelado                       |  R$ 4,00  |                 ')
    print('-------------------------------------------------------------------------------------------------------')
    print('Seja bem vindo {}, digite o código do que deseja pedir:'.format(nome))

#Na função valid, é realizada a validação da inserção do núemro do pedido do cliente,
# garantido que o cliente digite um valor válido de pedido, caso contrário, é exibida uma mensagem de erro,
# informando que o valor digitado não existe, e é exibido novamente a interface de inserção do número do pedido
#Nessa função, também é recolhido o número do pedido pela variável 'pedido'.
def valid ():
    vali=''
    while not vali:
        try:
            pedido = int(input('--->'))
        except ValueError:
            print ('Código selecionado não existente!')
            continue
        for i in range(99, 261):
            if pedido==i:
                vali ='ok'
        if vali != 'ok':
            print ('Código selecionado não existente!')
            continue
        return pedido

#Na função val, é realizada a verificação de qual o pedido realizado pelo usuário (produto e seu valor),
# através do número pedido gravada na variável 'pedido'.
# Essa verificação é realizada através do uso da operação if e gravada nas variáveis globais 'valor' e 'prod'.
def val(pedido):
    v=pedido
    global prod, valor
    if v == 100:
        valor=9.00
        prod='Cachorro Quente Simples'
    if v == 110:
        valor=11.00
        prod = 'Cachorro Quente Duplo'
    if v == 120:
        valor=13.00
        prod = 'X Salada'
    if v == 130:
        valor=12.00
        prod = 'X Egg'
    if v == 140:
        valor = 14.00
        prod = 'X Bacon'
    if v == 150:
        valor = 17.00
        prod = 'X Tudão'
    if v == 200:
        valor = 5.00
        prod = 'Refrigerante Lata'
    if v == 210:
        valor = 4.00
        prod = 'Chá Gelado'
    return valor



#Na função criarquivo, é criado um arquivo para gravar o pedido do cliente, com o nome dele inserido na função menu.
def criaarquivo(nomearquivo):
    try:
        a = open(nomearquivo, 'wt+')
        a.close()
    except:
        print ('Ops! Algo deu errado...')


#Na função gravarpedido, é gravado o produto e o valor pedido pelo usuário à cada novo pedido da sessão.
def gravarpedido(nomearquivo, produto, valor):
    try:
        a = open(nomearquivo, 'at')
    except:
        print('Ops! Algo deu errado...')
    else:
        a.write('Produto: {}                            Valor: {}\n'.format(produto, valor))
    finally:
        a.close()

#Na função lerpedido, é lido o arquivo do pedido do cliente, com as informações gravadas dos produtos pedidos e seus
#respectivos valores
def lerpedido(nomearquivo):
    try:
        a = open(nomearquivo, 'rt')
    except:
        print('Ops! Algo deu errado...')
    else:
        print (a.read())
    finally:
        a.close()

#Bloco gráfico inicial de boas vindas, onde é também solicitado o nome do usuário para a criação do arquivo de seu
#pedido
print ('*-----------------------------------------------------------------------------------------------------*')
print ('|                 BEM VINDO À LANCHONETE DA PATRÍCIA CANOSSA GAGLIARDI                                |')
print ('*-----------------------------------------------------------------------------------------------------*')
nome=input('Olá! Informe seu nome: ')
#Após isso, é criado o arquivo txt com seu nome, chamando a função criarquivo com o nome recebido no
# input com a variável 'nome'.
arquivo='{}.txt'.format(nome)
criaarquivo (arquivo)
#Após criado o arquivo, é chamada a função menu para apresentar o menu ao usuário, e o mesmo realizar o seu primeiro
#pedido de produto
menu()
#São criadas as variáveis fim, é utilizada na verificação de novos pedidos ou do fim do pedido, bem como, também é reco-
#lhido o valor e o produto pedido pelo usuário através da função val(pedido), onde o valor é guardo na varável 'valort'
# e o produto na 'prod'.
fim=''
prod=''
pedido=valid()
valort =float(val(pedido))
produto=prod
#Após, é gravado no arquivo do pedido, o produto e o valor do pedidd do usuário através da função gravarpedido
gravarpedido(arquivo, produto, valor)

#Nesse bloco de verificação, é verificado através de while, se o usuário deseja pedir mais produtos ou não.
#Para isso, inicia-se com a verificação da variável verificadora 'fim', que inicialmente esta vazia, sendo em uma
#verificação booleana uma 'False', e o while, continuando até a variável conter algum valor e se tornar 'True".
while not fim:
    print ('')
#è pedido pára o usuário informar se quer pedido mais algum produto, inserindo 'n' ou 's', sendo gravado esse valor
#na variável 'mais'
    mais=input('Deseja pedir mais alguma coisa? (s/n)->')
    #Através da função if, se verificado que o usuário digitou 's', o que significa que ele deseja pedir mais algum produto,
    #É novamente chamada a função valid, para gravar o número do pedido, recolhido o valor ('v') do pedido
    # e o produto ('prod') pela função val(pedido), e a variável 'valort', passa a ser somada ao valor recolhido
    #para gerar o valor total do pedido a ser exibido para o usuário.
    # [E também, gravado o novo produto e valor do mesmo no arquivo txt do pedido
    if mais=='s':
        pedido = valid()
        v=float(val(pedido))
        valort += v
        produto=prod
        gravarpedido(arquivo, produto, v)
    #Aqui, é verificado se o usuário digitou corretamente 's' ou 'n' na pergunta, Caso tenha digitado algo diferente,
    #é exibida mensagem de erro, e o usuário terá que digitar novamente se deseja mais algum produto ou não
    elif mais!='s' and mais!='n':
        print ('Opção inexistente...')
    #Nesse bloco, caso o usuário tenha digitado 'n', que seria para finalizar o pedido, a variável 'fim' recebeo valor
    #'fim', se tornando True. É informado ao usuário que o pedido foi finalizado, e exibido o pedido, através da
    #função lerpedido, que lê o arquivo txt com o pedido gravado, e também exibido o valor total do pedido, gravado
    #na variável valort,
    elif mais=='n':
        fim='fim'
        print ('Finalizando o pedido...\n')
        print('O seu pedido foi: ')
        lerpedido(arquivo)
        print ('| Valor Total: R${} |'.format(valort))
        #Ao final desse if, após a variável 'fim' se tornar True, o while é encerrado, e o programa finalizado.
```
